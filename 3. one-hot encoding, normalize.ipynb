```python
import pandas as pd
data = pd.read_csv('data/train_set/2.data_add_feature1.csv', header = 0)
data.columns
```

```python
## 1. 사용할 데이터만 가져오기
```

```python
data_use = data.drop(columns = ['방송일시', '상품명', 'date','상품코드','마더코드','time'])
data_use_fit = data_use
data_use_fit
```

## 2. 데이터의 종류
#### 정규화를 해야할 것 : '기온(°C)_seoul', '강수량(mm)_seoul', '습도(%)_seoul', '적설(cm)_seoul', '기온(°C)_busan', '강수량(mm)_busan', '습도(%)_busan', '적설(cm)_busan', 'weekday', 'hour',  '노출(분)','quater', 판매단가

#### 처리가 필요(one-hot) :  '상품군',  'time'

#### 그대로 사용 :  'pmWarning_seoul', 'pmWarning_busan', '휴일', 

#### 정답 : 취급액

#### standard_scalar_list

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

standard_scalar_list = ['기온(°C)_seoul', '기온(°C)_seoul',
                                                         '강수량(mm)_seoul', '습도(%)_seoul', '적설(cm)_seoul',
                                                         '기온(°C)_busan', '강수량(mm)_busan', '습도(%)_busan',
                                                         '적설(cm)_busan', 'weekday', 'hour', '노출(분)','quater', '판매단가']

data_use_fit[standard_scalar_list] = std_scaler.fit_transform(pd.DataFrame(data_use[standard_scalar_list]))
data_use_fit
```

```python
from sklearn.preprocessing import LabelEncoder
one_hot_list = [ '상품군']
encoder = LabelEncoder()
data_use_fit[one_hot_list]=encoder.fit_transform(data_use[one_hot_list])/10 
```
